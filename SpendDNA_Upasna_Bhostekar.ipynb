```python
# ==========================================================
# Parse Multiple Date Formats
# ==========================================================

from datetime import datetime     # using datetime library


def parse_date(date):

# Different date formats present in the dataset
# Try every possible date format one by one
    formats = [
        "%Y-%m-%d",     # 2024-01-01
        "%d-%b-%y",     # 01-Jan-24
        "%d %b %Y",     # 01 Jan 2024
        "%d/%m/%y"      # 01/01/24
    ]

    for fmt in formats:

        try:      # Try converting the date using the current format
            return datetime.strptime(date, fmt)        # converts string into proper datetime format

        except ValueError:      # handles error
            continue            # continue program

    return pd.NaT        # Return NaT if none of the formats match
```

```python
# ==========================================================
# Apply Date Parser
# ==========================================================

df_clean["Date"] = df_clean["Date"].apply(parse_date)

print("Date Parsing Completed ✅")
```

```python
def extract_vendor(description):

    description = description.upper().strip()

    # UPI Transactions
    if description.startswith("UPI-"):

        vendor = description[4:]           # Remove "UPI-"
        vendor = vendor.split("@")[0]      # Remove everything after '@'
        vendor = vendor.split("-")[0]      # Remove extra IDs after '-'
        return vendor

    # BHIM Transactions
    if description.startswith("BHIM-"):
        return description.replace("BHIM-", "")      # Removes 'BHIM-'

    if description.startswith("BHIM "):
        return description.replace("BHIM ", "")      # Removes 'BHIM'

    # POS Transactions
    if description.startswith("POS "):

        vendor = description.replace("POS ", "")      # Remove "POS "
        vendor = vendor.split("-")[0]                 # Remove extra words after '-'
        vendor = vendor.split()[0]                    # Keep only the vendor name
        return vendor


    # NEFT Transactions
    if description.startswith("NEFT-"):

        vendor = description.replace("NEFT-", "")        # Remove "NEFT-"
        vendor = vendor.split("-SALARY")[0]              # Remove salary information
        return vendor

    # IMPS Transactions
    if description.startswith("IMPS-"):

        vendor = description.replace("IMPS-", "")       # Remove "IMPS-"
        vendor = vendor.split("-")[1]                   # Extract the vendor name
        return vendor

    # Special Vendor Names
    # Handle vendors that appear in different formats

    if "SWIGGY" in description:
        return "SWIGGY"

    if "ZOMATO" in description:
        return "ZOMATO"

    if "BLINKIT" in description:
        return "BLINKIT"

    if "ZEPTO" in description:
        return "ZEPTO"

    if "UBER" in description:
        return "UBER"

    if "OLA" in description:
        return "OLA"

    if "AMAZON" in description:
        return "AMAZON"

    if "FLIPKART" in description:
        return "FLIPKART"

    if "BOOKMYSHOW" in description or "BMS" in description:
        return "BOOKMYSHOW"

    if "THIRDWAVE" in description or "TWC" in description:
        return "THIRD WAVE"

    if "STARBUCKS" in description:
        return "STARBUCKS"

    if "DMART" in description or "AVENUE SUPERMARTS" in description:
        return "DMART"

    if "GROFERS" in description:
        return "BIGBASKET"

    if "ANI TECHNOLOGIES" in description:
        return "OLA"

    if "BANGALORE ELEC SUPPLY" in description or "BESCOM" in description:
        return "BESCOM"

    # Default Vendor

    # If no rule matches, then return the first word of description
    return description.split()[0]

    # AI-assisted: Used as a learning aid to improve implementation and code comments.
```

```python
# ==========================================================
# Extract Vendors
# ==========================================================

df_clean["Vendor"] = df_clean["Description"].apply(extract_vendor)

print("Vendor Extraction is Done ✅")
```

📌 Feature 3 - Category Tagger

```python
# ==========================================================
# Category Tagger
# ==========================================================

def assign_category(vendor):

    # Convert vendor name to uppercase to ensure case-insensitive matching
    vendor = vendor.upper()

    # Let's categorize Food & Cafes
    if vendor in [
        "SWIGGY", "ZOMATO", "STARBUCKS", "TRUFFLES",
        "CCD", "CAFE", "COFFEE", "THIRDWAVE",
        "EMPIRE", "MEGHANA", "DINEOUT"
    ]:
        return "Food"

    # Let's categorize Grocery
    elif vendor in [
        "BLINKIT", "ZEPTO", "DMART",
        "BIGBASKET", "INSTAMART"
    ]:
        return "Grocery"

    # Let's categorize e-commmerce and fashion Shopping
    elif vendor in [
        "AMAZON", "AMAZONPAY",
        "FLIPKART", "MYNTRA", "NYKAA"
    ]:
        return "Shopping"

    # Let's categorize Transport services
    elif vendor in [
        "UBER", "OLA", "OLACABS",
        "BMTC", "ROPPEN", "RAPIDO",
        "TUMMOC-BMTC"
    ]:
        return "Transport"

    # Let's categorize Fuel
    elif vendor in [
        "HP", "BPCL", "INDIAN"
    ]:
        return "Fuel"

    # Let's categorize electricity and water bills Utilities
    elif vendor in [
        "BESCOM", "BWSSB"
    ]:
        return "Utilities"

    # Let's categorize Mobile Recharge / Bills
    elif vendor in [
        "AIRTEL", "JIORECHARGE", "VI"
    ]:
        return "Recharge"

    # Categorize movie, music and OTT platform expenses, etc. Entertainment
    elif vendor in [
        "BOOKMYSHOW", "NETFLIX",
        "DISNEY", "SPOTIFY"
    ]:
        return "Entertainment"

    # Let's Categorize investment related transactions
    elif vendor in [
        "GROWWPAY", "ZERODHA"
    ]:
        return "Investment"

    # Let's Categorize salary credits
    elif vendor == "TECHCRUSH LABS":
        return "Salary"

    # Let's Categorize money transferred to friends
    elif vendor in [
        "AMAN", "ANKIT", "PRIYA", "VIKAS"
    ]:
        return "Friend Transfer"

    # Let's Categorize ATM cash withdrawals
    elif vendor.startswith("ATM"):
        return "Cash Withdrawal"

    # Let's Assign "Others" if no category matches
    else:
        return "Others"
```

```python
# ==========================================================
# Create Category Column
# ==========================================================

df_clean["Category"] = df_clean["Vendor"].apply(assign_category)

print("✅ Category Tagging is Done")
```

📌 Feature 4 - Spending Overview

```python
# ==========================================================
# Spending Overview
# ==========================================================

# Filter debit transactions (money spent)
debit_transactions = df_clean[df_clean["Type"] == "debit"]

# Filter credit transactions (money received)
credit_transactions = df_clean[df_clean["Type"] == "credit"]

# Calculate total amount spent
total_debit = debit_transactions["Amount"].sum()

# Calculate total amount received
total_credit = credit_transactions["Amount"].sum()

# Calculate net balance = (Income - Expenses)
net_balance = total_credit - total_debit

# Calculate avg transaction amount
average_transaction = df_clean["Amount"].mean()

print("=" * 70)
print("SPENDING OVERVIEW")
print("=" * 70)

print(f"Total Debit          : ₹{total_debit:,.2f}")
print(f"Total Credit         : ₹{total_credit:,.2f}")
print(f"Net Balance          : ₹{net_balance:,.2f}")
print(f"Average Transaction  : ₹{average_transaction:,.2f}")

# AI-assisted: Assisted in spending analysis calculations.
```

```python
# ==========================================================
# Spending by Category
# ==========================================================

# Group all debit (expense) transactions by their category
# and calculate the total spending in each category
category_spending = (
    debit_transactions
    .groupby("Category")["Amount"]      # Group transactions based on Category
    .sum()                              # Calculate total spending for each category
    .sort_values(ascending=False)       # Sort categories from highest to lowest spending
)

# Display category-wise spending summary
print("=" * 70)
print("SPENDING BY CATEGORY")
print("=" * 70)

print(category_spending)
```

```python
# ==========================================================
# Top Vendors by Spending
# ==========================================================

# Group all debit (expense) transactions by vendor
# and calculate the total amount spent with each vendor
vendor_spending = (
    debit_transactions
    .groupby("Vendor")["Amount"]       # Group transactions based on Vendor
    .sum()                             # Calculate total spending for each vendor
    .sort_values(ascending=False)      # Sort vendors from highest to lowest spending
)

# Display the top 10 vendors with the highest spending
print("=" * 70)
print("TOP 10 VENDORS")
print("=" * 70)

print(vendor_spending.head(10))
```
